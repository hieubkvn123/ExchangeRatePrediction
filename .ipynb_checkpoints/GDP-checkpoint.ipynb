```python
import pandas as pd

### Loading the datasets ###
gdp = pd.read_csv('data/gdp.csv',header=0).dropna()
exchange_rate = pd.read_csv('data/exchange_rate.csv',header=0).dropna()



def discard_date(string):
    splits = string.split('/')
    year = splits[2]
    month = splits[1]
    half_year = 'H1' 
    
    if(int(year) <= 20):
        year = '20' + str(year) 
    else:
        year = '19' + str(year)
    
    if( int(month) > 6):
        half_year = 'H2'
        
    return year + '-' + half_year

def discard_me(string):
    splits = string.split('-')
    month = splits[1]
    half_year = 'H1'
    
    if((month) == Q1):
        half_year = 'H2'
    
    
gdp_mean = gdp.groupby('quarter').mean()

exchange_rate['date'] = exchange_rate['date'].apply(discard_date)
exchange_rate = exchange_rate.groupby('date').mean()

### change column from date -> half_year in exchange rate dataset ###
exchange_rate.index.names =['half_year']

### change column from value -> exchange_rate in exchange rate dataset ###
exchange_rate = exchange_rate.rename(columns={'value':'exchange_rate'})

gdp_mean['quarter'] = gdp['quarter'].apply(discard_me)

gdp_mean.index.names =['half_year']
gdp_mean = gdp_mean.rename(columns={'value':'GDP'})

print(exchange_rate, gdp_mean)
```

```python
import matplotlib.pyplot as plt

#plt.plot(exchange_rate['value'])
#plt.show()

fig, ax= plt.subplots(figsize=(15,8))
ax.plot(exchange_rate['exchange_rate'], label='Exchange Rate')

ax2 = ax.twinx()
ax2.plot(gdp_mean['GDP'], color='orange', label='GDP Value')

plt.legend()
plt.show()
```

```python
merged = pd.merge(exchange_rate, gdp_mean, on='quarter')
#print(merged)
merged.corr()
```
